```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
"""load iris data from sklearn library
insert the first to record of the data in X to use it as binary regression
insert the target data in y but if the value of trage is not 0 change it to 1 using '*1' 
"""
iris = datasets.load_iris()
X = iris.data[:, :2]
y = (iris.target != 0) * 1
print (X)
print(y)
```

```python
"""drow the figer size 5*5 
then scatter the data that have y = 0 with green color 
and the data that has the value of y = 1 with red
the legned function show the label the show the red represent 1 and 0 is green"""
plt.figure(figsize=(5, 5))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='g', label='0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='r', label='1')
plt.legend();
```
